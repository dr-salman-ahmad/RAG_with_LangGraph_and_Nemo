```python
import os

class ConfigManager:
    def __init__(self, base_path: str = 'config'):
        self.base_path = base_path
        self.rails_path = os.path.join(self.base_path, 'rails')
        self._setup_directories()

    def _setup_directories(self) -> None:
        """Create necessary directories if they do not exist."""
        if os.path.exists(self.base_path):
            os.system(f"rm -rf {self.base_path}")

        os.makedirs(self.base_path, exist_ok=True)
        os.makedirs(self.rails_path, exist_ok=True)

    def _get_file_path(self, file_name: str) -> str:
        """Return the file path for a given file name."""
        return os.path.join(self.base_path, file_name)

    def _get_rails_file_path(self, file_name: str) -> str:
        """Return the file path within the 'rails' folder for a given file name."""
        return os.path.join(self.rails_path, file_name)

    def create_and_write_file(self, file_path: str, content: str) -> None:
        """Create and write content to a file.

        Args:
            file_path (str): The path of the file to create and write to.
            content (str): The content to write to the file.
        """
        try:
            with open(file_path, 'w') as file:
                file.write(content)
            print(f"Successfully wrote to {file_path}")
        except Exception as e:
            print(f"Failed to write to {file_path}. Error: {e}")

    def initialize_files(self, config_yml_content: str, general_co_content: str,
                         prompts_yml_content: str, disallowed_co_content: str) -> None:
        """Initialize and write content to the necessary configuration files."""
        files_and_contents = {
            self._get_file_path('config.yml'): config_yml_content,
            self._get_file_path('general.co'): general_co_content,
            self._get_file_path('prompts.yml'): prompts_yml_content,
            self._get_rails_file_path('disallowed.co'): disallowed_co_content
        }

        for file_path, content in files_and_contents.items():
            self.create_and_write_file(file_path, content)

if __name__ == "__main__":
    config_manager = ConfigManager()
    config_manager.initialize_files(config_yml_content, general_co_content, prompts_yml_content, disallowed_co_content)
```

```python
import re
from typing import Annotated, Iterator, Literal, TypedDict, List, Optional

from langchain_openai import ChatOpenAI
from langchain_community.document_loaders import web_base
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
from langchain_core.output_parsers import StrOutputParser
from langchain_core.messages import BaseMessage, AIMessage, convert_to_messages
from langchain_core.pydantic_v1 import BaseModel, Field
from langchain_core.retrievers import BaseRetriever
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_anthropic import ChatAnthropic
from langchain_openai import OpenAIEmbeddings
from langgraph.graph import END, StateGraph, add_messages
from langgraph.checkpoint import MemorySaver
from langchain.memory import ConversationBufferMemory
from langchain.prompts import PromptTemplate, ChatPromptTemplate
from langchain.output_parsers.openai_functions import JsonKeyOutputFunctionsParser
from langchain.utils.openai_functions import convert_pydantic_to_openai_function
from nemoguardrails import LLMRails, RailsConfig
from IPython.display import Image, display

class CheckPerson(BaseModel):
    """Is name, phone number or email is present."""

    binary_score: bool = Field(
        description="Answer addresses the question, 'True' or 'False'"
    )

class CustomMemory:
    def __init__(self):
        self.conversation_memory = ConversationBufferMemory(
            memory_key="chat_history",
            return_messages=True
        )
        self.user_info = {}

    def save_context(self, inputs, outputs):
        self.conversation_memory.save_context(inputs, outputs)

    def load_memory_variables(self, inputs):
        return self.conversation_memory.load_memory_variables(inputs)

    def save_user_info(self, name, age=None, email=None):
        if name in self.user_info:
            if age is not None:
                self.user_info[name]['age'] = age
            if email is not None:
                self.user_info[name]['email'] = email
        else:
            self.user_info[name] = {}
            if age is not None:
                self.user_info[name]['age'] = age
            if email is not None:
                self.user_info[name]['email'] = email

    def print_user_info(self):
        for name, info in self.user_info.items():
            print(f"Name: {name}, Age: {info.get('age')}, Email: {info.get('email')}")

class Demographic(BaseModel):
    """Information about a person."""
    name: str = Field(description="Person's name")
    age: Optional[int] = Field(default=None, description="Person's age")
    email: Optional[str] = Field(default=None, description="Person's email address")


class Format(BaseModel):
    """Information to extract."""
    demographic: List[Demographic] = Field(description="List of info about people")

class GraphState(TypedDict):
    messages: Annotated[list[BaseMessage], add_messages]
    question: str
    documents: list[Document]
    candidate_answer: str

class DocumentProcessor:
    def __init__(self, persist_directory='docs/chroma'):
        self.persist_directory = persist_directory
        self.embedding = OpenAIEmbeddings()
        self.vectordb = None
        self.retriever = None

    def setup(self, documents):
        # Text Splitting
        splitter = RecursiveCharacterTextSplitter(
            chunk_size=100,
            chunk_overlap=5,
            separators=["(?<=\. )", " "]
        )
        splits = splitter.split_documents(documents)

        # Remove old database files if any
        if os.path.exists(self.persist_directory):
            os.system(f"rm -rf {self.persist_directory}")

        # Storing documents in ChromaDB
        self.vectordb = Chroma.from_documents(
            documents=splits,
            embedding=self.embedding,
            persist_directory=self.persist_directory
        )

        # Retrieving documents from ChromaDB
        self.retriever = self.vectordb.as_retriever(
            search_type="mmr",
            search_kwargs={'k': 2, 'fetch_k': 8}
        )
        self.vectordb.persist()

class ChatWorkflow:
    def __init__(self):
        self.model = ChatOpenAI(model_name="gpt-3.5-turbo", temperature=0)
        self.config = RailsConfig.from_path("config")
        self.llm = LLMRails(config=self.config, llm=self.model)
        self.retriever = None
        self.memory = CustomMemory()
        self.processor = DocumentProcessor()
        self.workflow = self.create_workflow()

    def format_messages(self, message, relevant_chunks):
        return [{"role": "context", "content": {"relevant_chunks": relevant_chunks}}, {"role": "user", "content": message}]

    def process_documents(self):
        documents = [
            Document(page_content="Devsloop is a software development firm based in Gujranwala, Pakistan, with a team "
                                  "of 20 employees. The company is led by CEO Ammad Javaid, with Shehwar Khalid as a "
                                  "partner. Devsloop has managed Adly clients and has worked on notable projects such "
                                  "as Beambox, Rapid Translate, Pageflows, Upcall, and Easylama"
                                  "There is Salman Ahmad. Salman is learning Data Science. He joined the company in "
                                  "September 2023. He did Graduation from Gift University"),
            Document(page_content="It has an HR manager named Emama Babur, and salaries are disbursed on the 1st of "
                                  "each month. Devsloop provide a machine either MAC or Laptop to all employees"),
            Document(page_content="Devsloop has foose ball game too.")
        ]
        self.processor.setup(documents)
        self.retriever = self.processor.retriever

    def document_search(self, state: GraphState):
        question = convert_to_messages(state["messages"])[-1].content
        documents = self.retriever.invoke(question)
        return {"documents": documents, "question": question}

    def generate(self, state: GraphState):
        question = state["question"]
        result = state["documents"]
        documents = '\n'.join([doc.page_content for doc in result])

        generation = self.llm.generate(messages=self.format_messages(question, documents))
        self.memory.save_context({"input": question}, {"output": generation["content"]})
        return {"candidate_answer": generation['content']}

    def finalize_response(self, state: GraphState):
        return {"messages": [AIMessage(content=state["candidate_answer"])]}

    def extract_information(self, state: GraphState):
        prompt = ChatPromptTemplate.from_messages([
            ("system", "Extract the relevant information; if not explicitly provided, do not guess. Extract partial info"),
            ("human", "{message}")
        ])
        extraction_functions = [convert_pydantic_to_openai_function(Format)]
        extraction_model = self.model.bind(functions=extraction_functions, function_call={"name": "Format"})
        extraction_chain = prompt | extraction_model | JsonKeyOutputFunctionsParser(key_name="demographic")
        extracted_info = extraction_chain.invoke({"message": state['question']})

        # Store or update extracted user information in custom memory
        for person in extracted_info:
            name = person.get('name')
            age = person.get('age')
            email = person.get('email')

            self.memory.save_user_info(name, age, email)
        return {"question": state['question']}

    def is_contain_user_info(self, state: GraphState):
        prompt = ChatPromptTemplate.from_messages([
            ("system", "Just check whether the question has person name, age or email; if yes then return True in boolean else return False"),
            ("human", "{question}")
        ])
        answer_grader = prompt | self.model.with_structured_output(CheckPerson)
        answer_grader: CheckPerson = answer_grader.invoke({"question": state['question']})
        return "extract_information" if answer_grader.binary_score else "generate"

    def create_workflow(self):
        workflow = StateGraph(GraphState)

        # Define the nodes
        workflow.add_node("document_search", self.document_search)
        workflow.add_node("generate", self.generate)
        workflow.add_node("finalize_response", self.finalize_response)
        workflow.add_node("extract_information", self.extract_information)

        # Build graph
        workflow.set_entry_point("document_search")
        workflow.add_edge("extract_information", "generate")
        workflow.add_edge("generate", "finalize_response")
        workflow.add_edge("finalize_response", END)

        workflow.add_conditional_edges(
            "document_search",
            self.is_contain_user_info
        )

        # Compile
        graph = workflow.compile()
        display(Image(graph.get_graph().draw_mermaid_png()))
        return graph

    def run(self, message):
        inputs = {"messages": [("human", message)]}
        for output in self.workflow.stream(inputs):
            print(output)
            print("\n---\n")

# Usage
chat_workflow = ChatWorkflow()
chat_workflow.process_documents()
chat_workflow.run("Hi There")
```
